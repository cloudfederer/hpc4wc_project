## evaluate simulations, extract arrays

```python
###64-1024 hard scaling
## get values of a4

nums_a4 = np.array([]) 
with open('a4_256_to1024.txt') as f:
    for line in f:
        line = line.split()       # check all ines in file separately
        if 'Elapsed' in line:     # check if string 'Elapsed' is present in the line
            nums_a4 = np.append(nums_a4, line[-2]).astype(float) # append second-last element (=number) to array
# numpy
a4_numpy_64 = nums_a4[0:7] # 7 items
a4_numpy_128 = nums_a4[7:14]
a4_numpy_256 = nums_a4[14:21]
a4_numpy_512 = nums_a4[21:28]
a4_numpy_1024 = nums_a4[28:35]
a4_numpy_weak = nums_a4[35:39] # 4 items
# ifirst
a4_ifirst_64 = nums_a4[39:46]
a4_ifirst_128 = nums_a4[46:53]
a4_ifirst_256 = nums_a4[53:60]
a4_ifirst_512 = nums_a4[60:67]
a4_ifirst_1024 = nums_a4[67:74]
a4_ifirst_weak = nums_a4[74:78]
# kfirst
a4_kfirst_64 = nums_a4[78:85]
a4_kfirst_128 = nums_a4[85:92]
a4_kfirst_256 = nums_a4[92:99]
a4_kfirst_512 = nums_a4[99:106]
a4_kfirst_1024 = nums_a4[106:113]
a4_kfirst_weak = nums_a4[113:117]
# gpu
a4_gpu_64 = nums_a4[117:124]
a4_gpu_128 = nums_a4[124:131]
a4_gpu_256 = nums_a4[131:138]
a4_gpu_512 = nums_a4[138:145]
a4_gpu_1024 = nums_a4[145:152]
a4_gpu_weak = nums_a4[152:156] 
# cuda
a4_cuda_64 = nums_a4[156:163]  
a4_cuda_128 = nums_a4[163:170]
a4_cuda_256 = nums_a4[170:177]
a4_cuda_512 = nums_a4[177:184]
a4_cuda_1024 = nums_a4[184:191]
a4_cuda_weak = nums_a4[191:195]


## get values of a1

nums_a1 =  np.array([])  
with open('a1_256_to1024.txt') as f:
    for line in f:
        line = line.split()       # check all ines in file separately
        if 'Elapsed' in line:     # check if string 'Elapsed' is present in the line
            nums_a1 = np.append(nums_a1, line[-2]).astype(float) # append second-last element (=number) to array
# numpy
a1_numpy_64 = nums_a1[0:7] # 7 items
a1_numpy_128 = nums_a1[7:14]
a1_numpy_256 = nums_a1[14:21]
a1_numpy_512 = nums_a1[21:28]
a1_numpy_1024 = nums_a1[28:35]
a1_numpy_weak = nums_a1[35:39] # 4 items
# ifirst
a1_ifirst_64 = nums_a1[39:46]
a1_ifirst_128 = nums_a1[46:53]
a1_ifirst_256 = nums_a1[53:60]
a1_ifirst_512 = nums_a1[60:67]
a1_ifirst_1024 = nums_a1[67:74]
a1_ifirst_weak = nums_a1[74:78]
# kfirst
a1_kfirst_64 = nums_a1[78:85]
a1_kfirst_128 = nums_a1[85:92]
a1_kfirst_256 = nums_a1[92:99]
a1_kfirst_512 = nums_a1[99:106]
a1_kfirst_1024 = nums_a1[106:113]
a1_kfirst_weak = nums_a1[113:117]
# gpu
a1_gpu_64 = nums_a1[117:124]
a1_gpu_128 = nums_a1[124:131]
a1_gpu_256 = nums_a1[131:138]
a1_gpu_512 = nums_a1[138:145]
a1_gpu_1024 = nums_a1[145:152]
a1_gpu_weak = nums_a1[152:156] 
# cuda
a1_cuda_64 = nums_a1[156:163]  
a1_cuda_128 = nums_a1[163:170]
a1_cuda_256 = nums_a1[170:177]
a1_cuda_512 = nums_a1[177:184]
a1_cuda_1024 = nums_a1[184:191]
a1_cuda_weak = nums_a1[191:195]
```
